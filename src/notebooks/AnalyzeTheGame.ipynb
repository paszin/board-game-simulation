# Setup for 3 players: How many cards per player?

```python
scores = {}
for card_count in range(3, 10):
    scores[card_count] = []
    random.seed(1892)
    for i in range(200):
        game = Game(players=[BasicPlayer(i) for i in range(1, 4)], cards_per_player=card_count)
        game.setup()
        game.simulate(quiet=True)
        # print("Remaining Cards:", game.score)
        scores[card_count].append(game.score)


```

```python
for card_count, score in scores.items():
    plt.plot(sorted(score), label=card_count)
plt.legend()
plt.title("Sorted Score for 3 player with N hand cards")
plt.show()
```
